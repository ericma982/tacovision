```python
import pandas as pd

tacos = pd.read_csv('final_final_tacos.csv')

tacos
```

```python
agg_tacos = []
cur_row_list = []
agg_columns = [
    'id', 
    'address',
    'categories',
    'city',
    'country',
    'cuisines',
    'latitude',
    'longitude',
    'menus.amountMax',
    'menus.amountMin',
    'name',
    'postalCode',
    'websites',
    'state',
    'priceRangeCategory',
    'restaurant_count',
    'taco_ratio',
    'burrito_ratio',
    'taco_meat_ratio',
    'taco_seafood_ratio',
    'taco_vegitarian_ratio',
    'taco_breakfast_ratio',
    'taco_other_ratio',
    'burrito_meat_ratio',
    'burrito_seafood_ratio',
    'burrito_vegetarian_ratio',
    'burrito_breakfast_ratio',
    'burrito_other_ratio'
]

index = 0
while index < len(tacos.index):
    first_row = tacos.loc[index]
    cur_row = first_row
    cur_row_list = [
        first_row['id'],
        first_row['address'],
        first_row['categories'],
        first_row['city'],
        first_row['country'],
        first_row['cuisines'],
        first_row['latitude'],
        first_row['longitude'],
        first_row['menus.amountMax'],
        first_row['menus.amountMin'],
        first_row['name'],
        first_row['postalCode'],
        first_row['websites'],
        first_row['state'],
        first_row['priceRangeCategory'],
        first_row['restaurant_count']
    ]
    
    category_count = {'taco_meat':0,
    'taco_seafood':0,
    'taco_vegetarian':0,
    'taco_breakfast':0,
    'taco_other':0,
    'burrito_meat':0,
    'burrito_seafood':0,
    'burrito_vegetarian':0,
    'burrito_breakfast':0,                  
    'burrito_other':0}
    taco_burrito = {'taco':0, 'burrito':0}
    total_menu_items = 0.0
    while(cur_row['id'] == first_row['id']):
        total_menu_items += 1
        str_index = ''
        if cur_row['type'] == 'taco':
            str_index = 'taco'
        else:
            str_index = 'burrito'
        taco_burrito[str_index] += 1
        
        str_index += '_' + cur_row['menus.type']
        category_count[str_index] += 1
        
        index += 1
        if index >= len(tacos.index):
            break
        cur_row = tacos.loc[index]
    cur_row_list.append(taco_burrito['taco'] / total_menu_items)
    cur_row_list.append(taco_burrito['burrito'] / total_menu_items)
    
    if(taco_burrito['taco'] > 0):
        cur_row_list.append(category_count['taco_meat'] / taco_burrito['taco'])
        cur_row_list.append(category_count['taco_seafood'] / taco_burrito['taco'])
        cur_row_list.append(category_count['taco_vegetarian'] / taco_burrito['taco'])
        cur_row_list.append(category_count['taco_breakfast'] / taco_burrito['taco'])
        cur_row_list.append(category_count['taco_other'] / taco_burrito['taco'])
    else:
        cur_row_list.append(0)
        cur_row_list.append(0)
        cur_row_list.append(0)
        cur_row_list.append(0)
        cur_row_list.append(0)
    
    if(taco_burrito['burrito'] > 0):
        cur_row_list.append(category_count['burrito_meat'] / taco_burrito['burrito'])
        cur_row_list.append(category_count['burrito_seafood'] / taco_burrito['burrito'])
        cur_row_list.append(category_count['burrito_vegetarian'] / taco_burrito['burrito'])
        cur_row_list.append(category_count['burrito_breakfast'] / taco_burrito['burrito'])
        cur_row_list.append(category_count['burrito_other'] / taco_burrito['burrito'])
    else:
        cur_row_list.append(0)
        cur_row_list.append(0)
        cur_row_list.append(0)
        cur_row_list.append(0)
        cur_row_list.append(0)
        
    agg_tacos.append(cur_row_list)
    

taco_places = pd.DataFrame(agg_tacos,columns=agg_columns)

taco_places
        
    
```
