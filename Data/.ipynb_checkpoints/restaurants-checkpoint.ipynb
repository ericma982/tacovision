```python
import pandas as pd

tacos = pd.read_csv('final_final_tacos.csv')

tacos
```

```python
agg_tacos = []
cur_row_list = []
agg_columns = [
    'id', 
    'address',
    'categories',
    'city',
    'country',
    'cuisines',
    'latitude',
    'longitude',
    'menus.amountMax',
    'menus.amountMin',
    'name',
    'postalCode',
    'websites',
    'state',
    'priceRangeCategory',
    'restaurant_count',
    'taco_ratio',
    'burrito_ratio',
    'meat_ratio',
    'seafood_ratio',
    'vegetarian_ratio',
    'breakfast_ratio',
    'other_ratio',
    'taco_meat_ratio',
    'taco_seafood_ratio',
    'taco_vegetarian_ratio',
    'taco_breakfast_ratio',
    'taco_other_ratio',
    'burrito_meat_ratio',
    'burrito_seafood_ratio',
    'burrito_vegetarian_ratio',
    'burrito_breakfast_ratio',
    'burrito_other_ratio'
]

index = 0
while index < len(tacos.index):
    first_row = tacos.loc[index]
    cur_row = first_row
    cur_row_list = [
        first_row['id'],
        first_row['address'],
        first_row['categories'],
        first_row['city'],
        first_row['country'],
        first_row['cuisines'],
        first_row['latitude'],
        first_row['longitude'],
        first_row['menus.amountMax'],
        first_row['menus.amountMin'],
        first_row['name'],
        first_row['postalCode'],
        first_row['websites'],
        first_row['state'],
        first_row['priceRangeCategory'],
        first_row['restaurant_count']
    ]
    
    category_count = {'taco_meat':0,
    'taco_seafood':0,
    'taco_vegetarian':0,
    'taco_breakfast':0,
    'taco_other':0,
    'burrito_meat':0,
    'burrito_seafood':0,
    'burrito_vegetarian':0,
    'burrito_breakfast':0,                  
    'burrito_other':0}
    taco_burrito = {'taco':0, 'burrito':0}
    total_menu_items = 0.0
    while(cur_row['id'] == first_row['id']):
        total_menu_items += 1
        str_index = ''
        if cur_row['type'] == 'taco':
            str_index = 'taco'
        else:
            str_index = 'burrito'
        taco_burrito[str_index] += 1
        
        str_index += '_' + cur_row['menus.type']
        category_count[str_index] += 1
        
        index += 1
        if index >= len(tacos.index):
            break
        cur_row = tacos.loc[index]
    cur_row_list.append(taco_burrito['taco'] / total_menu_items)
    cur_row_list.append(taco_burrito['burrito'] / total_menu_items)
    
    cur_row_list.append((category_count['taco_meat'] + category_count['burrito_meat']) / total_menu_items)
    cur_row_list.append((category_count['taco_seafood'] + category_count['burrito_seafood']) / total_menu_items)
    cur_row_list.append((category_count['taco_vegetarian'] + category_count['burrito_vegetarian']) / total_menu_items)
    cur_row_list.append((category_count['taco_breakfast'] + category_count['burrito_breakfast']) / total_menu_items)
    cur_row_list.append((category_count['taco_other'] + category_count['burrito_other']) / total_menu_items)
    
    if(taco_burrito['taco'] > 0):
        cur_row_list.append(category_count['taco_meat'] / taco_burrito['taco'])
        cur_row_list.append(category_count['taco_seafood'] / taco_burrito['taco'])
        cur_row_list.append(category_count['taco_vegetarian'] / taco_burrito['taco'])
        cur_row_list.append(category_count['taco_breakfast'] / taco_burrito['taco'])
        cur_row_list.append(category_count['taco_other'] / taco_burrito['taco'])
    else:
        cur_row_list.append(0)
        cur_row_list.append(0)
        cur_row_list.append(0)
        cur_row_list.append(0)
        cur_row_list.append(0)
    
    if(taco_burrito['burrito'] > 0):
        cur_row_list.append(category_count['burrito_meat'] / taco_burrito['burrito'])
        cur_row_list.append(category_count['burrito_seafood'] / taco_burrito['burrito'])
        cur_row_list.append(category_count['burrito_vegetarian'] / taco_burrito['burrito'])
        cur_row_list.append(category_count['burrito_breakfast'] / taco_burrito['burrito'])
        cur_row_list.append(category_count['burrito_other'] / taco_burrito['burrito'])
    else:
        cur_row_list.append(0)
        cur_row_list.append(0)
        cur_row_list.append(0)
        cur_row_list.append(0)
        cur_row_list.append(0)
        
    agg_tacos.append(cur_row_list)
    

taco_places = pd.DataFrame(agg_tacos,columns=agg_columns)

taco_places
        
    
```

```python
taco_places.to_csv('restaurants.csv')
```

```python
state_columns = [
    'state',
    'n_restaurants',
    'n_taco_restaurants',
    'n_burrito_restaurants',
    'taco_ratio',
    'burrito_ratio',
    'meat_ratio',
    'seafood_ratio',
    'vegetarian_ratio',
    'breakfast_ratio',
    'other_ratio',
    'taco_meat_ratio',
    'taco_seafood_ratio',
    'taco_vegetarian_ratio',
    'taco_breakfast_ratio',
    'taco_other_ratio',
    'burrito_meat_ratio',
    'burrito_seafood_ratio',
    'burrito_vegetarian_ratio',
    'burrito_breakfast_ratio',
    'burrito_other_ratio'
]

states = set()
states_n_rest = {}
states_n_tacos = {}
states_n_burritos = {}
state_taco = {}
state_burrito = {}
state_meat = {}
state_seafood = {}
state_vegetarian = {}
state_breakfast = {}
state_other = {}
state_taco_meat = {}
state_taco_seafood = {}
state_taco_vegetarian = {}
state_taco_breakfast = {}
state_taco_other = {}
state_burrito_meat = {}
state_burrito_seafood = {}
state_burrito_vegetarian = {}
state_burrito_breakfast = {}
state_burrito_other = {}

for index, row in taco_places.iterrows():
    if row['state'] not in states:
        states.add(row['state'])
        states_n_rest[row['state']] = 0
        states_n_tacos[row['state']] = 0
        states_n_burritos[row['state']] = 0
        state_taco[row['state']] = 0
        state_burrito[row['state']] = 0
        state_meat[row['state']] = 0
        state_seafood[row['state']] = 0
        state_vegetarian[row['state']] = 0
        state_breakfast[row['state']] = 0
        state_other[row['state']] = 0
        state_taco_meat[row['state']] = 0
        state_taco_seafood[row['state']] = 0
        state_taco_vegetarian[row['state']] = 0
        state_taco_breakfast[row['state']] = 0
        state_taco_other[row['state']] = 0
        state_burrito_meat[row['state']] = 0
        state_burrito_seafood[row['state']] = 0
        state_burrito_vegetarian[row['state']] = 0
        state_burrito_breakfast[row['state']] = 0
        state_burrito_other[row['state']] = 0
        
    states_n_rest[row['state']] += 1
    if row['taco_ratio'] > 0:
        states_n_tacos[row['state']] += 1
    if row['burrito_ratio'] > 0:
        states_n_burritos[row['state']] += 1
    state_taco[row['state']] += row['taco_ratio']
    state_burrito[row['state']] += row['burrito_ratio']
    state_meat[row['state']] += row['meat_ratio']
    state_seafood[row['state']]  += row['seafood_ratio']
    state_vegetarian[row['state']] += row['vegetarian_ratio']
    state_breakfast[row['state']] += row['breakfast_ratio']
    state_other[row['state']] += row['other_ratio']
    state_taco_meat[row['state']] += row['taco_meat_ratio']
    state_taco_seafood[row['state']] += row['taco_seafood_ratio']
    state_taco_vegetarian[row['state']] += row['taco_vegetarian_ratio']
    state_taco_breakfast[row['state']] += row['taco_breakfast_ratio']
    state_taco_other[row['state']] += row['taco_other_ratio']
    state_burrito_meat[row['state']] += row['burrito_meat_ratio']
    state_burrito_seafood[row['state']] += row['burrito_seafood_ratio']
    state_burrito_vegetarian[row['state']] += row['burrito_vegetarian_ratio']
    state_burrito_breakfast[row['state']] += row['burrito_breakfast_ratio']
    state_burrito_other[row['state']] += row['burrito_other_ratio']
    
state_rows = []
for state in states:
    print(states_n_rest[state])
    cur_row = [
        state,
        states_n_rest[state],
        states_n_tacos[state],
        states_n_burritos[state],
        state_taco[state] / states_n_rest[state],
        state_burrito[state] / states_n_rest[state],
        state_meat[state] / states_n_rest[state],
        state_seafood[state] / states_n_rest[state],
        state_vegetarian[state] / states_n_rest[state],
        state_breakfast[state] / states_n_rest[state],
        state_other[state] / states_n_rest[state],
        state_taco_meat[state] / states_n_tacos[state],
        state_taco_seafood[state] / states_n_tacos[state],
        state_taco_vegetarian[state] / states_n_tacos[state],
        state_taco_breakfast[state] / states_n_tacos[state],
        state_taco_other[state] / states_n_tacos[state],
        state_burrito_meat[state] / states_n_burritos[state],
        state_burrito_seafood[state] / states_n_burritos[state],
        state_burrito_vegetarian[state] / states_n_burritos[state],
        state_burrito_breakfast[state] / states_n_burritos[state],
        state_burrito_other[state] / states_n_burritos[state],
    ]
    state_rows.append(cur_row)
    
state_data = pd.DataFrame(state_rows,columns=state_columns)

state_data
```

```python
state_data.to_csv('state_data.csv')
```
